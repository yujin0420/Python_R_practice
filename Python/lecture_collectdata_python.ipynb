```python
# bike.csv 파일 읽기 및 통계량 산출
bike = open('bike.csv','r', encoding='utf-8')
idx=0
total=0
while True:
    line = bike.readline()
    if not line:
        break
    
    line = line.strip()
    if not line:
        continue
    
    if idx==0:     # 첫 행(변수명들) 건너뛰기
        idx=1
        continue
    
    cols = line.split(',')
    
    # 1. casual + registered = count 검증
    #casual = int(cols[-3])
    #registered = int(cols[-2])
    #count = int(cols[-1])
    
    #if (casual + registered) != count:
    #    print('Wrong: ', casual, registered, count)
    
    # 2. weather=1인 count 합계
    weather = cols[4]
    count = cols[-1]
    
    if count.isdigit():
        count = int(count)
    else:
        continue
    
    if weather == '1':
        total = total + count

bike.close()
print('Total: '+ str(total))
```
